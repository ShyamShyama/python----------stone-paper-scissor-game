```python
import random

options = [ "stone" , "paper" , "scissor" ]

computer_choice = random.choice(options)
user_choice = input("Enter your choice(stone/paper/scissor) :",).lower()

print("opponent choice is :",computer_choice)
print("your choice is :",user_choice)



if computer_choice == user_choice :
    print("Match draw")
elif (computer_choice == "stone" and user_choice == "scissor") or (computer_choice == "paper" and user_choice == "stone") or (computer_choice ==    "scissor" and user_choice == "paper") :
    print("Opponent wins")
else:
    print("You win")
```
